```python
from HuggingFaceParser import HuggingFaceParser
from huggingface_hub import hf_hub_url,get_hf_file_metadata,list_repo_files
a = HuggingFaceParser()
```

```python
paths
```

```python
total_size=0
for filename in paths:
    file_url = hf_hub_url(repo_id=repo_id,filename=filename)
    info = get_hf_file_metadata(file_url)
    total_size += info.size
for filename in paths_data:
    file_url = hf_hub_url(repo_id=repo_id,filename=filename)
    info = get_hf_file_metadata(file_url)
    total_size += info.size
```
